```python
from pathlib import Path
import pandas as pd
import tannson_fnc

# Extract files from applicable folders
base_folder = Path(r'D:\INVOICE')
sub_folder = Path(r'D:\INVOICE\ACTIVE')

all_files = list(base_folder.glob('*.xls')) + list(sub_folder.glob('*.xls'))

# Load extracted files
o_list = [] # List to store each file's order details
m_list = [] # List to store each file's monetary details
skip = list(range(8)) + list(range(12, 19)) # Skip initial label and customer info rows

i = 0 # File tracker
for file in all_files:
    # Make File # = Row Index for instances requiring manual inspection of files
    # Print output excluded to protect customer privacy but saved to txt file
    print(f"{i}. {file}")
    
    df = pd.read_excel(file, usecols=[1, 3], skiprows=skip, header=None, names=['Text', 'Numeric'], dtype=str)
    df = df.dropna(how='all')
    
    o_details, m_details = tannson_fnc.assign_data(df, file)
    m_details['File Type'] = file.name[0]
    
    o_list.append(o_details)
    m_list.append(m_details)
    
    i += 1

o_df = pd.concat(o_list, ignore_index=True)
m_df = pd.concat(m_list, ignore_index=True)

# base_df.to_csv(r'C:\Users\shuju\Desktop\Tannson_Tech_Project\raw_data.csv', index=False)
o_df.to_json(r'C:\Users\shuju\shus_workspace\tannson_proj\order_details.json', orient='records', indent=2, date_format='iso')
m_df.to_json(r'C:\Users\shuju\shus_workspace\tannson_proj\monetary_amts.json', orient='records', indent=2, date_format='iso')
```
